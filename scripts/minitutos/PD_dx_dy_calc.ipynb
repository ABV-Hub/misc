# Recalculate dx and dy in a dataframe by calling function

Built for one of the Shell Kuala Lumpur hackathon projects.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.random.randint(0,100,size=(5, 4)), columns=list('ABCD'))
df
```

```python
df['wellname'] = ['Well1', 'Well2', 'Patricia', 'Well4', 'Well5']
df
```

```python
df.rename(index=str, columns={"A": "Lat", "B": "Long"}, inplace=True)
df
```

```python
df.drop(columns=['C', 'D'], inplace=True)
df
```

```python
df
```

```python
df.loc[df['wellname'] == 'Patricia'].Lat.ndim
```

```python
def make_dx_dy(wellname):
    """updates df for new well"""
    df['dx'] = abs(df.Lat - df.loc[df['wellname'] == wellname].Lat.values)
    df['dy'] = abs(df.Long - df.loc[df['wellname'] == wellname].Long.values)
    
    return df
```

```python
make_dx_dy('Well1')
```

```python
make_dx_dy('Patricia')
```
