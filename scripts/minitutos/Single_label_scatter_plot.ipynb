# Single label in scatter plot

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = np.random.random(100)
y = np.random.random(100)
```

```python
x_lab, y_lab = 0.2, 0.8
```

```python
fig, ax = plt.subplots(figsize=(10,10), ncols=1)

ax.scatter(x, y)
ax.text(x_lab, y_lab, s='WellName', fontsize=16)
```
